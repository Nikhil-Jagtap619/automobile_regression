```python
# lets find the outliers
for col in df._get_numeric_data():
  plt.figure(figsize=(10,5))
  sns.boxplot(df[col])
  plt.title(col)
  plt.show()
```

```python
# fixing outliers
def fix_outlier(df,cols):
  for col in cols:
    quant1 = df[col].quantile(0.25)
    quant3 = df[col].quantile(0.75)
    iqr = quant3 - quant1
    up_limit = quant3 + 1.5*iqr
    low_limit = quant1 - 1.5*iqr
    df.loc[df[col] < low_limit, col] = low_limit
    df.loc[df[col] > up_limit, col] = up_limit
  return df
```

```python
# assigning the fixed df to df2
df2 = fix_outlier(df,['normalized-losses', 'wheel-base', 'length', 'width', 'engine-size', 'stroke','horsepower', 'highway-mpg','city-mpg', 'compression-ratio','price'])
```

```python
def fill_mean(df,cols):
  for col in cols:
    df[col] = pd.to_numeric(df[col], errors="coerce")
    df[col].fillna(round(df[col].mean(),2), inplace=True)
  return df
```

```python
# model building
from sklearn.model_selection import train_test_split, GridSearchCV, cross_val_score
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from lightgbm import LGBMRegressor
```

```python
# split into x and y variable
x = df_2.drop("price",axis=1)
y = df_2[["price"]]
```

```python
models = [LinearRegression(), Lasso(), Ridge(), ElasticNet(), KNeighborsRegressor(), DecisionTreeRegressor(), LGBMRegressor()]
```

```python
# a function to execute all the regression models at once
def reg_model(x,y, models):
  result_without_param = []
  for model in models:
    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=5)
    reg = model
    reg.fit(x_train, y_train)
    y_pred = reg.predict(x_test)
    print(model)
    print(f"Mean Squared Error: {mean_squared_error(y_test, y_pred)}\nR-Squared: {r2_score(y_test, y_pred)}")
    print("-"*45)
    result_without_param.append({'model': model, 
                        'MSE': mean_squared_error(y_test, y_pred),
                        'R2 Score': r2_score(y_test, y_pred)})
  return result_without_param
```

```python
output = reg_model(x, y, models)
```

```python
model_parameters={
    'Linear Regression':{
        'model': LinearRegression(),
        'params': {'fit_intercept': [True , False],
                   'normalize': [True, False], 
                    }},
    'Lasso Regression': {
        'model': Lasso(),
        'params': {'selection': ['cyclic', 'random'], 
                  'alpha': [0.1,0.01, 0.005, 0.05, 0.001 ,0.2, 0.3, 0.5, 0.8, 0.9, 1]}},
    'Ridge Regression' : {
        'model': Ridge(),
        'params': {'alpha': [0.1, 0.01, 0.005, 0.05, 0.001 ,0.2, 0.3, 0.5, 0.8, 0.9, 1], 
                   'solver': ['auto', 'svd', 'cholesky']}},
    'ElasticNet': {
        'model': ElasticNet(), 
        'params': {'alpha': [0.1,0.01, 0.005, 0.05, 0.001 ,0.2, 0.3, 0.5, 0.8, 0.9, 1],
                  'selection': ['cyclic', 'random']}},
    'KNeighborsRegeressor': {
        'model': KNeighborsRegressor(), 
        'params': {'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute'], 
                  'weights': ['uniform', 'distance'],
                  'n_neighbors': [3, 4, 5]}},
    'DecisionTreeRegressor': {
        'model': DecisionTreeRegressor(),
        'params': {'criterion': ['mse', 'friedman_mse', 'mae', 'poisson'],
                   'splitter': ['best', 'random']}}, 
    'LGBMRegressor': {
        'model': LGBMRegressor(),
        'params': {'learning_rate': [0.1, 0.2], 'max_depth': list(range(-1, 2, 1)), 
                   'n_estimators': list(range(97, 103))}}
}
```

```python
def tuning(x, y, model_params):
    results=[]
    for model_name, mp in model_params.items():
        X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.30, random_state=50)
        clf=GridSearchCV(mp['model'], mp['params'], cv=5, return_train_score=True)
        clf.fit(X_train, Y_train)
        Y_test_pred=clf.predict(X_test)
        mse=round(mean_squared_error(Y_test, Y_test_pred),4)
        r2score=round(r2_score(Y_test, Y_test_pred),4)                                                  
        print(model_name)
        print(f"Mean Squared Error: {mse}\nR-Squared: {r2score}")
        print('-'*45)
        results.append({'model': model_name, 
                        'MSE': mse,
                        'R2 Score': r2score,
                        'best_score': clf.best_score_,
                        'best_params': clf.best_params_})
    return results
```

```python
results = tuning(x,y,model_parameters)
```

```python
results = pd.DataFrame(results)
results = results.sort_values("R2 Score",ascending=False)
results
```
